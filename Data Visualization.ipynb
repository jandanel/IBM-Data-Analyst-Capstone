## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
import matplotlib.pyplot as plt

QUERY = """
SELECT *
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

```python
# histogram of ConvertedComp
sns.histplot(df.ConvertedComp)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
sns.boxplot(x = 'Age', data = df)
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
sns.scatterplot(x = 'Age', y = 'WorkWeekHrs', data = df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
plt.scatter(x = df.WorkWeekHrs, y = df.CodeRevHrs)
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)
db_df = df.DatabaseDesireNextYear.value_counts().head(5).reset_index()
db_df1 = db_df.rename(columns={'index':'Database', 'DatabaseDesireNextYear':'Count'})
plt.pie(x = db_df1.Count, labels = db_df1.Database, autopct='%.2f%%');
```

```python
# Pie chart of desired languages for next year
QUERY = """
SELECT * 
FROM LanguageDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)
lg_df = df.LanguageDesireNextYear.value_counts().reset_index()
lg_df1 = lg_df.rename(columns={'index':'Language', 'LanguageDesireNextYear':'Count'})
plt.pie(x = lg_df1.Count, labels = lg_df1.Language, autopct='%.2f%%');
```

```python
# currently working with SQL
QUERY = """
SELECT * 
FROM LanguageWorkedWith
"""

df = pd.read_sql_query(QUERY, conn)
sum(df.LanguageWorkedWith=='SQL')
```

```python
# Work with MySQL only
QUERY = """
SELECT * 
FROM DatabaseWorkedWith
"""

df = pd.read_sql_query(QUERY, conn)
df_mysql = df[df.DatabaseWorkedWith=='MySQL']
df_mysql_resp = df_mysql.Respondent.unique()

df_nomysql = df[df.DatabaseWorkedWith!='MySQL']
df_nomysql_resp = df_nomysql.Respondent.unique()

mysql_more = len(set(df_nomysql_resp).intersection(df_mysql_resp)) 
mysql_only = len(df_mysql_resp) - mysql_more
mysql_only
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT * 
FROM master
"""

df = pd.read_sql_query(QUERY, conn)
df_age = df[(df.Age >= 30) & (df.Age <= 35)]
df_age.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().plot.bar(stacked=True)
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_age2 = df[(df.Age >= 45) & (df.Age <= 60)]
df_age2.groupby('Age')['ConvertedComp'].median().plot.line()
```

```python
# Plot the median ConvertedComp for all ages from 25 to 30.
df_age2 = df[(df.Age >= 25) & (df.Age <= 30)]
df_age2.groupby('Age')['ConvertedComp'].median().plot.line()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
df.groupby('MainBranch')['MainBranch'].count().plot(kind='barh', figsize=(10, 6))
```
